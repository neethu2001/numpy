```python
#using matplotlib library we can pot the trigonometric function
x=np.arange(1,11)# x from 1 to 10
y=np.arange(10,110,10) # y from 10 to 110 with the gap of 10
plt.figure(figsize=(6,6))
plt.plot(x,y,'r--')
plt.show()
```

**plotting Trigonometric Curves**

```python
x_sin=np.arange(0,2*np.pi,0.1) #in x axis= 0 to 2pi with 0.1
y_sin=np.sin(x_sin) # all the sin values of x sin
print(y_sin)
```

```python

plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin)
plt.title('sin curve')
plt.show()
```

```python
x_cos=np.arange(0,2*np.pi,0.1)
y_cos=np.cos(x_cos)
plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos)
plt.title('cos curve')
plt.show()
```

```python
x_tan=np.arange(0,2*np.pi,0.1)
y_tan=np.tan(x_tan)

plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan)
plt.title('tan curve')
plt.show()
```

```python
x_cot=np.arange(0,2*np.pi,0.1)
y_cot=1/np.tan(x_cot)

plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot)
plt.title('cot curve')
plt.show()
```

**subplot**

```python
plt.figure(figsize=(6,6))

plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,'r-') # red dash line
plt.title('sin curve')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'b-') # blue dash line
plt.title('cos curve')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,'g-') # green dash line
plt.title('tan curve')

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot,'w-') # white dash line
plt.title('cot curve')

plt.show()
```
